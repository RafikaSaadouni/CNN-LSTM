```python
data_train = data_train[data_train['class'].isin(['DoS', 'Probe', 'R2L', 'normal'])]
data_test = data_test[data_test['class'].isin(['DoS', 'Probe', 'R2L', 'normal'])]
label_num = len(data_test['class'].unique())
label_num
```

```python
df_train_obj = data_train.iloc[:, :-4].select_dtypes(include='object')
df_train_num = data_train.iloc[:, :-4].select_dtypes(exclude='object')

print(f"shape of numeric features: {df_train_num.shape}")
print(f"shape of object features: {df_train_obj.shape}")
```

```python
df_test_obj = data_test.iloc[:, :-4].select_dtypes(include='object')
df_test_num = data_test.iloc[:, :-4].select_dtypes(exclude='object')

print(f"shape of numeric features: {df_test_num.shape}")
print(f"shape of object features: {df_test_obj.shape}")
```

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore')
# df_train_obj = df_train.iloc[:,:-3]
df_train_enc = enc.fit_transform(df_train_obj).toarray()
train_enc_features = enc.get_feature_names(input_features=df_train_obj.columns)
df_test_enc = enc.transform(df_test_obj).toarray()
test_enc_features = enc.get_feature_names(input_features=df_test_obj.columns)
# print(len(train_enc_features), len(test_enc_features))
X_train_enc = np.c_[df_train_num, df_train_enc]
X_test_enc = np.c_[df_test_num, df_test_enc]
print(f"X_train_enc shape: {X_train_enc.shape}")
print(f"X_test_enc shape: {X_test_enc.shape}")
```

```python
scaler = MinMaxScaler()
X_train_scaler = scaler.fit_transform(X_train_enc)
X_test_scaler = scaler.transform(X_test_enc)
X_train_normal = X_train_scaler[data_train['class']=='normal']
X_train_DoS = X_train_scaler[data_train['class']=='DoS']
X_train_Probe = X_train_scaler[data_train['class']=='Probe']
#X_train_U2R = X_train_scaler[df_train['class']=='U2R']
X_train_R2L = X_train_scaler[data_train['class']=='R2L']

X_test_normal = X_test_scaler[data_test['class']=='normal']
X_test_DoS = X_test_scaler[data_test['class']=='DoS']
X_test_Probe = X_test_scaler[data_test['class']=='Probe']
#X_test_U2R = X_test_scaler[df_test['class']=='U2R']
X_test_R2L = X_test_scaler[data_test['class']=='R2L']
```

```python
print("Normal samples in Training dataset:", X_train_normal.shape)
print(f"DoS samples in Training dataset: {X_train_DoS.shape}")
print(f"Probe samples in Training dataset: {X_train_Probe.shape}")
print(f"R2L samples in Training dataset: {X_train_R2L.shape}")
```

```python
print("Normal samples in Test dataset:", X_test_normal.shape)
print(f"DoS samples in Test dataset: {X_test_DoS.shape}")
print(f"Probe samples in Test dataset: {X_test_Probe.shape}")
print(f"R2L samples in Test dataset: {X_test_R2L.shape}")
```

```python
X_train_extract = X_train_scaler.copy()
X_test_extract = X_test_scaler.copy()
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train_extract = le.fit_transform(data_train['class'])
y_test_extract = le.transform(data_test['class'])

X_train_extract
```

```python
from sklearn import metrics
from tensorflow import keras
look_back = X_train_extract.shape[1]
def multiClassModel(units=4, look_back=look_back):
    model = Sequential()
    model.add(Input(shape=(1, look_back)))
    model.add(LSTM(units=50))
    #model.add(Dropout(0.3))
    model.add(Dense(label_num, activation="softmax"))
    model.compile(loss="sparse_categorical_crossentropy", optimizer='Adam' ,metrics=)
    model.summary()
    return model
```

```python
model = multiClassModel()
X_train_lstm = X_train_extract.reshape(X_train_extract.shape[0], 1, X_train_extract.shape[1])
X_test_lstm = X_test_extract.reshape(X_test_extract.shape[0], 1, X_test_extract.shape[1])
history = model.fit(X_train_lstm, y_train_extract, \
                   epochs=30, batch_size=64, validation_split=0.2)
### check the loss trend of epochs
pd.DataFrame(history.history).plot(kind='line', xlabel='epochs', figsize=(8, 6))
plt.show()
```
